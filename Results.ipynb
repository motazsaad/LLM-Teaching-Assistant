```python
import pandas as pd
```

```python
dataset = pd.read_csv('new_ds.csv')
```

```python
from sklearn.metrics import precision_score, recall_score, f1_score

# Ensure the columns are numeric
dataset["mistral_score"] = pd.to_numeric(dataset["mistral_score"], errors="coerce")
dataset["holistic_essay_score"] = pd.to_numeric(dataset["holistic_essay_score"], errors="coerce")

# Define a threshold to classify predictions (e.g., 3.0 for this example)
threshold = 3.0
dataset["predicted_class"] = (dataset["mistral_score"] >= threshold).astype(int)
dataset["predicted_class"]
dataset["actual_class"] = (dataset["holistic_essay_score"] >= threshold).astype(int)

# Calculate Precision, Recall, and F1 Score
precision = precision_score(dataset["actual_class"], dataset["predicted_class"], zero_division=1)
recall = recall_score(dataset["actual_class"], dataset["predicted_class"], zero_division=1)
f1 = f1_score(dataset["actual_class"], dataset["predicted_class"], zero_division=1)

# Print results
print(f"Precision: {precision:.2f}")
print(f"Recall: {recall:.2f}")
print(f"F1 Score: {f1:.2f}")
```

```python
import math
total = 0
true_grade = 0
for index, row in dataset.iterrows():
    mistral_score = row["mistral_score"]
    actual_score  = row["holistic_essay_score"]
    grade = mistral_score - actual_score
    if grade == 1:
      true_grade +=1
    total += 1
print(true_grade/total)
```

```python
from sklearn.metrics import cohen_kappa_score
from sklearn.metrics import mean_absolute_error, mean_squared_error
from scipy.stats import pearsonr, spearmanr
```

```python
df_specific_cleaned = dataset.dropna(subset=['mistral_score'])

mistral_scores = df_specific_cleaned["mistral_score"].values
human_scores   = df_specific_cleaned["holistic_essay_score"].values
mae = mean_absolute_error(human_scores, mistral_scores)
rmse = np.sqrt(mean_squared_error(human_scores, mistral_scores))
deviations = np.abs(np.array(mistral_scores) - np.array(human_scores))
avg_deviation = np.mean(deviations)

human_scores =  human_scores.astype(np.int64)
mistral_scores = mistral_scores.astype(np.int64)

kappa = cohen_kappa_score(mistral_scores, human_scores)

pearson_corr, _ = pearsonr(mistral_scores, human_scores)
spearman_corr, _ = spearmanr(mistral_scores, human_scores)


print(f"Mistral scores")


print(f"Mean Absolute Error (MAE): {mae}")
print(f"Root Mean Squared Error (RMSE): {rmse}")
print(f"Deviations from human scores: {deviations}")
print(f"Average absolute deviation: {avg_deviation}")
print(f"Pearson Correlation: {pearson_corr}")
print(f"Spearman Correlation: {spearman_corr}")

print(f"Cohen's Kappa: {kappa:.2f}")
```
