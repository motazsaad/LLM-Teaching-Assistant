```python
!pip install -U lightrag[ollama]

from lightrag.core.generator import Generator
from lightrag.core.component import Component
from lightrag.core.model_client import ModelClient
from lightrag.components.model_client import OllamaClient, GroqAPIClient
```

```python
import time


qa_template = r"""<SYS>
You are an expert teacher.
</SYS>
User: {{input_str}}
You:"""

class SimpleQA(Component):
    def __init__(self, model_client: ModelClient, model_kwargs: dict):
        super().__init__()
        self.generator = Generator(
            model_client=model_client,
            model_kwargs=model_kwargs,
            template=qa_template,
        )

    def call(self, input: dict) -> str:
        return self.generator.call({"input_str": str(input)})

    async def acall(self, input: dict) -> str:
        return await self.generator.acall({"input_str": str(input)})

from lightrag.components.model_client import OllamaClient
from IPython.display import Markdown, display
```

```python
models_client = []

for model in models_list:
  model_client = {
    "model_client": OllamaClient(),
    "model_kwargs": {"model": model}
  }
  models_client.append(model_client)
```

```python
new_asap = persaude.sample(n=20)
new_asap
```

```python
for model in models_client:
  print("Model Name:: "+model["model_kwargs"]["model"])
  total_time = 0
  for x , row in new_asap.iterrows():
    start = time.time()
    question_text = row["assignment"].split(' ')
    question_text_count = len(question_text)

    answer_text = row["full_text"].split(' ')
    answer_text_count = len(answer_text)




    new_prompt = (
      "You are an experienced English teacher with experience in argumentative essay tasked with grading a student essay. "
      "Assess the response critically and objectively. Your evaluation should focus solely on the quality of the answer in relation to the assignment's requirements.\n\n"
      "- Provide a total score from 1 to 6  based on the Lead , Position , Counterclaim , Rebuttal , Evidence ,and Concluding Summary of the response.\n"
      "The assignment details are as follows:\n"
      "- Question:"+ row["assignment"]+"\n"
      "- Student's Answer:+"+ row["full_text"] +"\n"
      "- Prompt Name: "+row["prompt_name"]+"\n\n"
      "gave me the final grade as  number only"
      "PLEASE PLEASE PLEASE DON'T GAVE ME ANY EXPLINATION"
      )


    modelQA = SimpleQA(**model)

    output = modelQA(new_prompt)
    end = time.time()
    new_asap.loc[x,"question_word_count"] = question_text_count
    new_asap.loc[x,"answer_word_count"] = answer_text_count
    new_asap.loc[x,model["model_kwargs"]["model"]+"_score"] = output.data

    total = end-start
    total_time += total
    new_asap.loc[x,model["model_kwargs"]["model"]+"_generation_time"] = total

    print("question_word_count",question_text_count)
    print("answer_word_count",answer_text_count)
    print("The Model Score is::",output.data)
    print("The dataset score is:",row["holistic_essay_score"])
    print("Time Generation:",end-start,"\n")


  avarage_time = total_time/len(new_asap)
  print("avarage time for qwen2 1.5b",avarage_time)
```
